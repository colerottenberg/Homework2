## Problem 3

Compute the 99\% confidence interval (CI) of the Bootstrap statistic for the NLSY79 dataset for NET_WEALTH.

Now we are interested in answering the question whether **baccalaureate education** is associated with an increase in **median net family wealth**?

1. Make sure you clean the data and separate them into two subsets based on whether the HIGHEST_GRADE_EVER is **larger than 12**. Report the **number of samples** and **median** of each dataset. 
2. To answer this question, build a Python function to implement CI. Consider having input parameters such as number of simulations and confidence value.
3. Using your function, resample a few times and see if the confidence interval varies. How many simulations are needed to get a reliable estimate of the 99% confidence interval for this statistic?
4. Discuss whether the confidence interval supports the hypothesis that baccalaureate is education associated with an increase in median net family wealth?.

```python
import pandas as pd
df = pd.read_csv('https://www.fdsp.net/data/nls.csv')
remap = {'R0000100' : 'CASE_ID',
         'T5597600' : 'GENDER',
         'T5684500' : 'NET_WEALTH',
         'T9900000' : 'HIGHEST_GRADE_EVER'
        }
df.rename(columns=remap,inplace=True)
df.head()
```

```python
# We need to clean the net wealth and highest grade ever columns to make sure they do not equal -5 or -3 as these are indicators of missing data
df = df.replace(-5, np.nan)
df = df.replace(-3, np.nan)
df = df.dropna(subset=['NET_WEALTH', 'HIGHEST_GRADE_EVER'])
highest_grade_12 = df.query("HIGHEST_GRADE_EVER <= 12 & HIGHEST_GRADE_EVER > 0")
past_12 = df.query("HIGHEST_GRADE_EVER > 12")
highest_grade_12.head()
past_12.head()
```

```python
import numpy as np

def bootstrap_CI(data, num_simulations, confidence=0.99):
    medians = []
    for _ in range(num_simulations):
        resample = np.random.choice(data, size=len(data), replace=True)
        medians.append(np.median(resample))
        
    lower = (1 - confidence) / 2
    upper = confidence + lower
    return (np.percentile(medians, 100 * lower),
            np.percentile(medians, 100 * upper))
```

```python
# Start with a few simulations and increase until CI stabilizes
for sims in [10, 100, 500, 1000, 5000, 10000]:
    ci = bootstrap_CI(past_12['NET_WEALTH'], sims, 0.99)
    print(f"99% CI with {sims} simulations: {ci}")
```
